# Import relevant packages

`sys`, `matplotlib`, `networkx`, `pandas`, `copy`, `itertools` and `pylab` are loaded to the workspace. 

```python
#!/usr/bin/env python

import sys
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import copy
import itertools
import pylab as pyt
```

# Construct edgelist

Edges are specified in an array.

```python
lines = ['1 2','1 5','1 6',
         '2 3','2 4','2 5','2 6','2 7',
         '3 4','3 7', 
         '4 5','4 7','4 8',
         '5 6','5 8',
         '6 7',
         '8 9',
         '9 10']
```

# Set labels

Labels are set for each node.

```python
labeldict={}
labeldict[1] = "Ed"
labeldict[2] = "Diane"
labeldict[3] = "Carol"
labeldict[4] = "Fernando"
labeldict[5] = "Garth"
labeldict[6] = "Beverly"
labeldict[7] = "Andre"
labeldict[8] = "Heather"
labeldict[9] = "Ike"
labeldict[10] = "Jane"
```

# Parse edgelist

The edgelist is parsed.

```python
G = nx.parse_edgelist(lines,nodetype = int)
```

# Fix positions of nodes

Positions for nodes are specified and plotted on the Cartesian plane.

```python
fixed_positions = {1:(1,1),
                   2:(1,3),
                   3:(1,5),
                   4:(2,4),
                   5:(2,2),
                   6:(0,2),
                   7:(0,4),
                   8:(3,3),
                   9:(4,3),
                   10:(5,3)}#set positions
fixed_nodes = fixed_positions.keys()
pos = nx.spring_layout(G,pos=fixed_positions, fixed = fixed_nodes)
```

# Plot the network

Figure parameters are set so that the plot extends to both margins.

The network is rendered; node size is specified to allow for labels to reside within each boundary. Options are set to further customize the plot.

```python
fig = plt.figure(figsize=(20,20))
ax = plt.subplot(111,aspect = 'equal')
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

nx.draw(G, pos, labels=labeldict, with_labels = True,node_size=6000, node_color="white", edgecolors="black")

plt.show()
```
