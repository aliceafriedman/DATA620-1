### Import relevant packages

`sys`, `matplotlib`, `networkx`, `pandas`, `copy`, `itertools` and `pylab` are loaded to the workspace. 

```python
import sys
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import copy
import itertools
import pylab as pyt
```

### Construct edgelist

Edges are specified in an array. By listing all edges in one line of code, relationships between nodes are easy to track.

```python
lines = ['1 2','1 5','1 6',
         '2 3','2 4','2 5','2 6','2 7',
         '3 4','3 7', 
         '4 5','4 7','4 8',
         '5 6','5 8',
         '6 7',
         '8 9',
         '9 10']
```

### Set labels

Labels are set for each node; the integer-valued index of each node is used to define the label.

```python
labeldict={}
labeldict[1] = "Ed"
labeldict[2] = "Diane"
labeldict[3] = "Carol"
labeldict[4] = "Fernando"
labeldict[5] = "Garth"
labeldict[6] = "Beverly"
labeldict[7] = "Andre"
labeldict[8] = "Heather"
labeldict[9] = "Ike"
labeldict[10] = "Jane"
```

### Parse edgelist

The edgelist is parsed using `parse_edgelist`.

```python
G = nx.parse_edgelist(lines,nodetype = int)
```

### Fix positions of nodes

Positions for nodes are specified and plotted on the Cartesian plane.

```python
fixed_positions = {1:(1,1),
                   2:(1,3),
                   3:(1,5),
                   4:(2,4),
                   5:(2,2),
                   6:(0,2),
                   7:(0,4),
                   8:(3,3),
                   9:(4,3),
                   10:(5,3)}#set positions
fixed_nodes = fixed_positions.keys()
pos = nx.spring_layout(G,pos=fixed_positions, fixed = fixed_nodes)
```

### Plot the network

Figure parameters are set so that the plot extends to both margins.

The network is rendered; node size is specified to allow for labels to reside within each boundary. Options are set to further customize the plot.

- `with_labels` includes specified labels in the plot
- `node_size` sets node size
- `node_color` sets color of each node
- `edgecolors` sets color of relationships drawn between nodes
- `font_size` sets size of font
- `width` specifies edge width

The initial excerpt of code suppresses a deprecation warning from matplotlib; further documentation and information for this error can be found at https://es.stackoverflow.com/questions/268663/deprecatedwarning-cb-iterable-al-usar-networkx.

```python
import warnings
from matplotlib.cbook.deprecation import MatplotlibDeprecationWarning

warnings.filterwarnings("ignore", category=MatplotlibDeprecationWarning)

fig = plt.figure(figsize=(20,20))
ax = plt.subplot(111,aspect = 'equal')
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

nx.draw(G, pos, labels=labeldict, 
        with_labels = True, 
        node_size=15000, 
        node_color="#ECE0C6", 
        edgecolors="black", 
        font_size=20,
        width=10)

plt.show()
```
